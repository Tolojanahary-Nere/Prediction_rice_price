```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Charger dataset
df = pd.read_csv("../data/prix_riz_madagascar.csv")

# Statistiques simples
print(df.groupby("region")["price"].mean())

# Courbe
df.groupby("date")["price"].mean().plot()
plt.title("Prix moyen du riz à Madagascar")
plt.show()

# Prédiction simple
X = df.index.values.reshape(-1, 1)
y = df["price"]
model = LinearRegression().fit(X, y)
print("Prix prévu prochain:", model.predict([[len(X)+1]])[0])
```
